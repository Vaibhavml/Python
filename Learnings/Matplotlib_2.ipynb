```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"])
plt.xlabel('Sepal Length')
plt.ylabel("Petal Length")
```

```python
iris["Species"] = iris["Species"].replace({"Iris-setosa":0, "Iris-versicolor":1,"Iris-virginica":2})
# Giving different species diffrent coding
iris.sample(5)
```

```python
# Giving colour encoding to the different species of flower
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"], c = iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel("Petal Length")

plt.show()
```

```python
# Giving colour bar to the different species of flower
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"], c = iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel("Petal Length")
plt.colorbar()

plt.show()
```

```python
# Defining the opacity to our graph
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"], c = iris['Species'],alpha = 0.8)
plt.xlabel('Sepal Length')
plt.ylabel("Petal Length")
plt.colorbar()

plt.show()
```

```python
# To increase the graph size
plt.figure(figsize=(12,6))

plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"], c = iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel("Petal Length")
plt.colorbar()

plt.show()
```

```python
sample_df
```

```python
plt.scatter(sample_df["avg"],sample_df["strike_rate"])
```

```python
# To give names to each sample dots we will give text to these dots
# Example
x = [2,4,6,8]
y = [10,20,30,40]
plt.scatter(x,y)

plt.text(2,10,"Point 1")
plt.text(4,20,"Point x")
plt.text(6,30,"Point y")
plt.text(8,40,"Point 4", fontdict = {'size':12,"color":"green"})
```

```python
# We will do the same thing in this data
plt.scatter(sample_df["avg"],sample_df["strike_rate"])

for i in range(sample_df.shape[0]):
  plt.text(sample_df["avg"].values[i],sample_df["strike_rate"].values[i],sample_df["batter"].values[i])
```

```python
# To increase the redability
plt.figure(figsize = (16,8))
plt.scatter(sample_df["avg"],sample_df["strike_rate"])

for i in range(sample_df.shape[0]):
  plt.text(sample_df["avg"].values[i],sample_df["strike_rate"].values[i],sample_df["batter"].values[i])
```

```python
# We want to increase the dot size as per runs scored by a batsman
plt.figure(figsize = (16,8))
plt.scatter(sample_df["avg"],sample_df["strike_rate"],s = sample_df["runs"])

for i in range(sample_df.shape[0]):
  plt.text(sample_df["avg"].values[i],sample_df["strike_rate"].values[i],sample_df["batter"].values[i])
```

### Draw Horizontal and vertical lines on a graph

```python
plt.figure(figsize = (16,8))
plt.scatter(sample_df["avg"],sample_df["strike_rate"],s = sample_df["runs"])

plt.axhline(130,color = "red")# For Horizontal line

plt.axvline(30,color = "green") # For Vertical Line

for i in range(sample_df.shape[0]):
  plt.text(sample_df["avg"].values[i],sample_df["strike_rate"].values[i],sample_df["batter"].values[i])
```

## 3D Surface Plot

```python
x =np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
np.meshgrid(x,y)
```

```python
xx,yy = np.meshgrid(x,y)
```

```python
xx.shape
```

```python
z1 = xx**2 + yy**2
```

```python
z1.shape
```

```python
fig = plt.figure(figsize = (12,8))

ax = plt.subplot(projection = "3d")

graph = ax.plot_surface(xx,yy,z1,cmap = "viridis")

fig.colorbar(graph)
```

```python
#Playing With Different function
z= np.sin(xx)+np.cos(yy)

fig = plt.figure(figsize = (12,8))

ax = plt.subplot(projection = "3d")

graph = ax.plot_surface(xx,yy,z,cmap = "viridis")

fig.colorbar(graph)
```

## Contour Plot
 - It shows the 3D plot on the 2D surface

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z1,cmap='viridis')
fig.colorbar(p)
```
