```python
# Q: You have to create a new column named "Euclidian distance" with apply
# Basically you have to work with row
def euclidian(row):
  pt_A = row['1st point']
  pt_B = row['2st point']

  return ((pt_A[0]-pt_B[0])**2 + (pt_A[1]-pt_B[1])**2)**0.5
```

```python
points_df.apply(euclidian, axis = 0)
points_df
```

```python
points_df['distance'] = points_df.apply(euclidian,axis=1)
points_df
```
