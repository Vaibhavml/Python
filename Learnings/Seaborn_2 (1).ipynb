## Categorical Plots

- Box Plot:
A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

```python
# Box Plot
sns.boxplot(data= tips, x= 'sex', y = 'total_bill')
```

```python
sns.boxplot(data= tips, x= 'day', y = 'total_bill', color = "red")
```

```python
# Using catplot
sns.catplot(data= tips, x= 'day', y = 'total_bill', kind = 'box')
```

```python
# Using hue parameter
# Using catplot
sns.catplot(data= tips, x= 'day', y = 'total_bill', kind = 'box', hue = 'sex')
```

```python
# Using axes level
sns.boxplot(data= tips, x= 'day', y = 'total_bill', hue = 'sex')
```

```python
# If we want to see the overall plot of single column
sns.boxplot(data = tips, y ='total_bill')
```

## Count plot

- A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable

```python
sns.countplot(data = tips,x = 'sex', hue = 'day')
```

```python
# Faceting using boxplot
sns.catplot(data = tips, x= 'sex',y = 'total_bill', col = 'smoker',kind = 'box')
```

```python
sns.catplot(data = tips, x= 'sex',y = 'total_bill', col = 'smoker',kind = 'box', row = 'time')
```

## Regplot:
- Regresson Plot is a function in seaborn which not only plots a scatter plot but also tries to give a best fit line.

```python
tips.head()
```

```python
sns.scatterplot(data = tips, x ='total_bill', y = 'tip' )
```

```python
# Since we want the regression line also so we will use reg plot
sns.regplot(data = tips, x ='total_bill', y = 'tip' )
# axes level function

# Hue parameter is not available
```

##In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
#lm plot
sns.lmplot(data = tips, x ='total_bill', y = 'tip' , hue = 'sex')
```

```python
#resid plot
sns.residplot(data = tips,x='total_bill', y ='tip')
```

## Residual plot gives the possitive error or residu in the linear regeression line above the zero line and it gives the negative error below the zero line
- It besically denotes the error in the data with respect to the linear regression line

```python
## Second way to plot facet plot ->> facet grid
# figure level -> relplot -> displot -> catplot -> lmplot
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
```

## With same x and y axis element it is plotting kde plot

```python
## Pair Grid
sns.PairGrid(data = iris)

#This will give us a object
```

```python
g = sns.PairGrid(data = iris)
# g.map

g.map(sns.scatterplot)
```

```python
# Adding hue parameter
g = sns.PairGrid(data = iris, hue = 'species')
# g.map

g.map(sns.scatterplot)
```

```python
# We can plot diagonal items and offdioganal item differently
g =sns.PairGrid(data = iris)
g.map_diag(sns.histplot)
```

```python
# We can plot diagonal items and offdioganal item differently
g =sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# We can plot diagonal items and offdioganal item differently
g =sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.scatterplot)
```

```python
## If we want to plot kde plot
g =sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot)
```

```python
## If we want to plot box plot
g =sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)
```

```python
# We can customize our graph as per our need

g =sns.PairGrid(data = iris, hue = 'species')

g.map_diag(sns.histplot) # In diagonal -> Histogram
g.map_upper(sns.kdeplot) # In upper part -> Kde
g.map_lower(sns.scatterplot) # In lower part -> scatter
```
