## plt.plot Vs plt.scatter plot

- If you have huge dataset then in that case you can plot scatter plot using plt.plot (faster)
- If you have a small dataset and you want to show a scatter plot then u can use plt.scatter

```python
# scatterplot using plt.plot
# faster
plt.plot(tips['total_bill'],tips['tip'],'o')
```

## Bar Graph

- It is used for Bivariate Analysis
- Numerical Vs Categorical Columns
- Use Case - Aggregate Analysis of Groups

```python
# Simple Bar Chart
children = [10,20,70,40,60]
colour = ['red','blue','green','yellow','pink']
plt.bar(colour,children)
```

```python
#Giving different colours
# Simple Bar Chart
children = [10,20,70,40,60]
colour = ['red','blue','green','yellow','pink']
plt.bar(colour,children, color = "Green")
```

#We can provide all the labels to these charts

```python
# To make Horizontal bar charts
children = [10,20,70,40,60]
colour = ['red','blue','green','yellow','pink']
plt.barh(colour,children, color = "Green") # Use plt.barh
```

```python
#Importing Batsman_season_record
df = pd.read_csv("/content/batsman_season_record.csv")
df.head(5)
```

```python
plt.bar(df['batsman'],df["2016"])
```

```python
# Plotting multiple bar plot in matplotlib
plt.bar(df['batsman'],df["2016"],width = 0.5)
```

```python
# To plot three consecutive year's bar graph in a single bar graph
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# We may face a problem if the label name is too lengthy
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
```

#Here we can see that the name on the x axis are overlapping, to avoid this we are using xtics

```python
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation=70)
```

```python
# We can play with the rotation angles
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation="vertical")
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

##Piechart

```python
# Simple Piechart
data = [25,40,60,85,95]
plt.pie(data)

plt.show()
```

```python
# To make it informative
data = [25,40,60,85,95]
subject = ["Hindi","English","math","sst","csc"]
plt.pie(data,labels = subject)

plt.show()
```
