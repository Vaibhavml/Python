```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
```

```python
tips  = sns.load_dataset('tips')
```

```python
# Scatter plot is axes level function
# It appears square in shape
sns.scatterplot(data = tips, x = "total_bill", y = "tip")
```

```python
# relplot is figure level function
# It appears in square shape
sns.relplot(data = tips, x = "total_bill", y = "tip",kind = "scatter")
```

```python
# adding hue parameter
sns.scatterplot(data = tips, x = "total_bill", y = "tip",hue = "sex")
```

```python
# Adding different parameters
# adding hue parameter
sns.scatterplot(data = tips, x = "total_bill", y = "tip",hue = "sex", style = "time")
```

```python
# adding size parameter
sns.scatterplot(data = tips, x = "total_bill", y = "tip",hue = "sex", style = "time",size = "size")
```

## KDE Plot

```python
tips = sns.load_dataset('tips')
```

```python
sns.kdeplot(data = tips, x = 'total_bill')
```

```python
sns.displot(data = tips, x ='total_bill', kind= 'kde', hue = 'sex' )
```

```python
# Fill parameter
sns.displot(data = tips, x ='total_bill', kind= 'kde', hue = 'sex', fill =True )
```

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.displot(data = tips, x ='total_bill', kind= 'kde' )
```

```python
sns.kdeplot(data = tips, x = 'total_bill')
```

```python
sns.rugplot(data = tips, x = 'total_bill')
```

```python
sns.kdeplot(data = tips, x = 'total_bill')
sns.rugplot(data = tips, x = 'total_bill')

# It is used as complementry graph with other graphs
```

```python

# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

sns.histplot(data=tips, x='total_bill', y='tip')
```

```python
#  Same task can be performed by this also
sns.displot(data=tips, x='total_bill', y='tip',kind = 'hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian function
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

## Heatmap
- It is color encoded matrix

```python
gap = px.data.gapminder()
gap.head()
```

# This is long dataset, we have to convert it into wide dataset

```python
gap.pivot(index = 'country', columns = 'year', values = 'lifeExp')
```

```python
temp_df = gap.pivot(index = 'country', columns = 'year', values = 'lifeExp')
```

```python
sns.heatmap(temp_df)
```

```python
# To increase the figure size we can use the matplot lib option
plt.figure(figsize = (15,10))
sns.heatmap(temp_df)
```

```python
gap[gap['continent']=='Europe'].pivot(index = 'country',columns = "year", values = 'lifeExp')
```

```python
temp_df = gap[gap['continent']=='Europe'].pivot(index = 'country',columns = "year", values = 'lifeExp')
sns.heatmap(temp_df)
```

```python
# To get the life expetency value
# We are using annotation
temp_df = gap[gap['continent']=='Europe'].pivot(index = 'country',columns = "year", values = 'lifeExp')

plt.figure(figsize = (15,15))
sns.heatmap(temp_df, annot = True)
```

```python
temp_df = gap[gap['continent']=='Europe'].pivot(index = 'country',columns = "year", values = 'lifeExp')

plt.figure(figsize = (15,15))
sns.heatmap(temp_df, annot = True,linewidth = 0.5)
```

```python
# To change the colour shade
temp_df = gap[gap['continent']=='Europe'].pivot(index = 'country',columns = "year", values = 'lifeExp')

plt.figure(figsize = (15,15))
sns.heatmap(temp_df, annot = True,linewidth = 0.5, cmap = "viridis")
```

```python
# To change the colour shade
temp_df = gap[gap['continent']=='Europe'].pivot(index = 'country',columns = "year", values = 'lifeExp')

plt.figure(figsize = (15,15))
sns.heatmap(temp_df, annot = True,linewidth = 0.5, cmap = "summer")
```
