```python
missing_value_per = df.isnull().sum()/df.shape[0] * 100
missing_value_per
```

```python
# To know which columns have missing values
missing_value_clm_gre_20 = missing_value_per[missing_value_per > 20].keys()
missing_value_clm_gre_20
```

```python
df2_drop_clm = df.drop(columns=missing_value_clm_gre_20)
df2_drop_clm.shape
```

```python
# We have to see only those columns which has numerical data
df3_num=df2_drop_clm.select_dtypes(include=['int64','float64'])
df3_num.head()
```

```python
# Now geberate the heat map
plt.figure(figsize=(16,9))
sns.heatmap(df3_num.isnull())
```

```python
# To know about the row in which we have null values
df3_num[df3_num.isnull().any(axis=1)]
```

```python
df3_num.isnull().sum()
```

```python
# To know about the columns in which we have missing values
missing_num_var = [var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
missing_num_var
```

Here we have seen that some columns have null values and we need to put some values into them, to do that we will be needing mean/median values, but before that we will have to see the data distribution of those columns

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth':5, 'color':'#DC143C'})
```

1) If the variation of data is normal distribution, then in that case we can
use Mean/Median/Mode any of those to fill the missing values.
                                                                               
2) If the variation of data is Right/Left skewed, then in that case we use Median to fill the missing values.
Note : **The data points should be randomly distributed**




```python
df4_num_mean = df3_num.fillna(df3_num.mean())
df4_num_mean.isnull().sum().sum()
```

```python
# We will now compare the data distribution of original data and the data set which we have cleaned and filled with mean values
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df4_num_mean[var], bins=20, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    plt.legend()
```



1.  Now we have seen the variation in data when we have used **mean** to fill the NA values.
2.   Now we will use **median** to fill the NA values and then we will see the distribution



```python
# Here we have filled the null values with median
df5_num_median = df3_num.fillna(df3_num.median())
df5_num_median.isnull().sum().sum()
```

```python
# Here we have taken the original data value, the data value where we used mean, and then the values where we used median and see the distribution
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20,hist=False, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df4_num_mean[var], bins=20,hist=False, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    sns.distplot(df5_num_median[var], bins=20,hist=False, kde_kws={'linewidth':3, 'color':'k'},label="Median",)
    plt.legend()
```
