For the better vizualization and seperate column name

```python
pd.concat([df["MSZoning"].value_counts()/df.shape[0]*100,
          df3_dropped_rows["MSZoning"].value_counts()/df3_dropped_rows.shape[0]*100], axis = 1,
           keys=["MSZoning_org", "MSZoning_clean"])
```

```python
df.shape
```

```python
# We will define a function to show the variation in the categorical data
# You only need to place the category name and you will be able to check it's distplot/ distribution
def categorical_variable_distribution(var):
  return pd.concat([df[var].value_counts()/df.shape[0]*100,
         df3_dropped_rows[var].value_counts()/df3_dropped_rows.shape[0]*100], axis = 1,
         keys=[var+'_org', var+"_clean"])
```

```python
categorical_variable_distribution('MSZoning')
```

```python
categorical_variable_distribution('Street')
```

```python
categorical_variable_distribution('LotShape')
```
